```python
import pandas as pd
from sklearn import svm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/LearnPythonWithRune/MachineLearningWithPython/main/files/gender.csv')
data.head()
```

```python
data["class"]=data["Gender"].apply(lambda x: "r" if x=="F" else "b")
```

```python
data.head()
```

```python
data=data.iloc[:100]
```

```python
fig, ax =plt.subplots()
ax.scatter(x=data["Height"],y=data["Hair length"],c=data["class"])
ax.axhline(20)
import numpy as np
x=np.array([150,190])
ax.plot(x,x-145,c="b")
ax.plot(x,2*x-310,c="g")
```

```python
ax.axhline(20)
```

```python
X=data[["Height","Hair length"]]
Y=data["Gender"]
Y=np.array([0 if Gender=="M" else 1 for Gender in Y])
```

```python
clf=svm.SVC(kernel="linear")
clf.fit(X,Y)
```

```python
X_test=np.random.rand(10000,2)
X_test=X_test*(50,120)+(150,0)

Y_pred=clf.predict(X_test)
```

```python
fig, ax = plt.subplots()

ax.scatter(x=X_test[:,0], y=X_test[:,1], c=Y_pred, alpha=.25)
Y_color = ['r' if value == 0 else 'b' for value in Y]
ax.scatter(x=X['Height'], y=X['Hair length'], c=Y_color)
```

```python
import numpy as np
import pandas as pd
from sklearn import svm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/LearnPythonWithRune/MachineLearningWithPython/main/files/data_02.csv')
data.head()
```

```python
class_ids = {'Dobermann': 0, 'German Shepherd': 1, 'Rottweiler': 2}
```

```python
data["class_id"]=data["Class"].apply(lambda x:class_ids[x])
data.head()
```

```python
ig, ax =plt.subplots()
ax.scatter(x=data["Weight"],y=data["Height"],c=data["Result"])

```

```python
clf=svm.SVC(kernel="linear")
clf.fit(data[["Weight","Height"]],data["Result"])
```

```python
X_test=np.random.rand(10000, 2)
X_test=X_test*(40, 20) + (25, 55)
y_pred=clf.predict(X_test)
```

```python
fig, ax = plt.subplots()
ax.scatter(x=X_test[:,0], y=X_test[:,1], c=y_pred, alpha=.25)
colors=["b","r","g"]
ax.scatter(x=data["Weight"], y=data["Height"], c=data["Result"].apply(lambda x: colors[x]))
```
