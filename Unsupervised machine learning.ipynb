```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=np.random.randn(400,2)
data[:100] += 5,5
data[100:200] +=10,10
data[200:300] +=10,5
data[300:] += 5,10
```

```python
fig, ax=plt.subplots()

ax.scatter(x=data[:,0], y=data[:,1])
```

```python
model = KMeans(n_clusters=4, init='random', random_state=42, max_iter=20, n_init=1)

model.fit(data)

y_pred = model.predict(data)

fig, ax = plt.subplots()
ax.scatter(x=data[:,0], y=data[:,1], c=y_pred)
ax.scatter(x=model.cluster_centers_[:,0], y=model.cluster_centers_[:,1], c='r')
```

```python
model = KMeans(n_clusters=10, init='random', random_state=42, max_iter=20, n_init=1)

model.fit(data)

y_pred = model.predict(data)

fig, ax = plt.subplots()
ax.scatter(x=data[:,0], y=data[:,1], c=y_pred)
ax.scatter(x=model.cluster_centers_[:,0], y=model.cluster_centers_[:,1], c='r')
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=np.random.randn(400,2)
data[:100] += 5,5
data[100:200] +=10,10
data[200:300] +=10,5
data[300:] += 5,10
```

```python
sse=[]
for k in range (1,11):
    model=KMeans(n_clusters=k)
    model.fit(data)
    sse.append(model.inertia_)
```

```python
fig, ax=plt.subplots()

ax.plot(range(1,11),sse)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silouette_coefficients=[]
for k in range (2,11):
    model = KMeans(n_clusters=k)
    model.fit(data)
    score=silhouette_score(data,model.labels_)
    silouette_coefficients.append(score)
```

```python
fig, ax=plt.subplots()

ax.plot(range(2,11),silouette_coefficients)
```
